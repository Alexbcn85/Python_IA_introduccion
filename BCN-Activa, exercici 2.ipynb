## Spark SQL

RDD, Dataframes i Datasets

```
import os.path
import re
import os
import pandas as pd
from matplotlib import pyplot as plt
from math import floor
from pyspark import SparkConf, SparkContext, SQLContext, HiveContext
from pyspark.sql import Row
from pyspark.sql.functions import desc
fileName = "dbfs:/FileStore/tables/tweets/tweets28a_sample.json"
print(fileName)
```

```
tweets_sample = sqlContext.read.json(fileName)
```

```
tweets_sample.count()
```

```
tweets_sample.take(1)
```

```
tweets_sample.printSchema()
```

```
tweets_rdd=tweets_sample.rdd
```

```
tweets_rdd.take(1)
```

```
tweets_fields=tweets_rdd.map(lambda item: (item[0], item[1]))
```

```
tweets_fields.take(1)
```

```
sqlContext.registerDataFrameAsTable(tweets_sample, "tweets_sample")
```

```
tweets1 = sqlContext.sql("""  SELECT *
                                FROM tweets_sample
                                WHERE user.lang == 'es'""")
```

```
tweets1.show()
```

```
tweets2 = sqlContext.sql("""  SELECT user
                                FROM tweets_sample
                                WHERE user.lang == 'es'""")
```

```
tweets2.show(10,False)
```

```
users_agg = sqlContext.sql("""  SELECT user.screen_name, COUNT(*) AS tweets
                                FROM tweets_sample
                                WHERE user.lang == 'es'
                                GROUP BY user.screen_name
                                ORDER BY tweets DESC""")
```

```
users_agg.show()
```

```
users = tweets_sample.where("user.lang == 'es'").select("user.screen_name")
```

```
users.show(10,False)
```

```
users_agg = tweets_sample.where("user.lang == 'es'").select("user.screen_name").groupBy("screen_name")\
                 .agg({"screen_name":"count"})\
                 .orderBy("count(screen_name)", ascending=False)
users_agg.limit(10).show()
```

```
tweets_filtered=tweets_sample.select("user.screen_name", "created_at", "user.followers_count", "text").orderBy(desc("user.followers_count"))
tweets_filtered.show()
```
